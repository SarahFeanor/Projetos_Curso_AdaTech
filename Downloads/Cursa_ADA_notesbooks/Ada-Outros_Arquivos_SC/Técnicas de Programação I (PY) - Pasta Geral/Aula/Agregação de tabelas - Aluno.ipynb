### Merge

##### Exemplo 1

```python
df_customer = pd.DataFrame({
 'id':[1, 2, 3, 4],
 'name':['Tom','Jenny', 'James','Dan']   
})

df_info = pd.DataFrame({
    'id':[2, 3, 4, 5],
    'age':[31, 20, 40, 70],
    'sex':['F','M','M','F']
})
```

```python
df_customer
```

##### Modos de mesclar

![](https://s3-sa-east-1.amazonaws.com/lcpi/910ed9dc-2768-47c3-9f21-ade9e156eaa7.png)

**1. inner**: mescla linhas cujas chaves estejam em ambos os dataframes.<br>
**2. outer**: mescla todas as linhas de ambos os dataframes.<br>
**3. left**: mescla todas as linhas de y que estiverem contidas em x.<br>
**4. right**: mescla todas as linhas de x que estiverem contidas em y.<br>

#### 1. inner

![](https://s3-sa-east-1.amazonaws.com/lcpi/00764540-d97a-496a-96e7-e0418da993a4.png)

```python
df_inner = df_customer.merge(df_info, on='id', how='inner')
df_inner
```

#### 2. outer

![](https://s3-sa-east-1.amazonaws.com/lcpi/25e34804-6980-41a4-a95b-1628c6ecc6f6.png)

```python
df_outer = df_customer.merge(df_info, on='id', how='outer')
df_outer
```

#### 3. left

![](https://s3-sa-east-1.amazonaws.com/lcpi/d6fd1e43-7e7c-405c-983c-434e224bb868.png)

```python
df_left = df_customer.merge(df_info, on='id', how='left')
df_left
```

#### 4. right

![](https://s3-sa-east-1.amazonaws.com/lcpi/d3349192-e04a-4d9b-b3a0-8ff9c82f8bd4.png)

```python
df_right = df_customer.merge(df_info, on='id', how='right')
df_right
```

3. Entre solteiros e casados, quem conseguiu mais empréstimos?

```python
df[df['aprovacao_emprestimo'] == 'Y']['estado_civil'].value_counts()
```

```python
#Sim = Casado 
#Não = Solteiro

qtd_solteiro = ((df['estado_civil']=='No') & (df['aprovacao_emprestimo']=='Y')).sum()
qtd_casado = ((df['estado_civil']=='Yes') & (df['aprovacao_emprestimo']=='Y')).sum()
print(qtd_casado, qtd_solteiro)

if qtd_solteiro > qtd_casado:
    print(f'A quantidade de solteiros é maior com {qtd_solteiro}')
elif qtd_solteiro < qtd_casado:
    print(f'A quantidade de casados é maior com {qtd_casado}')
else:
    print(f'A quantidade de casados e solteiros é a mesma com {qtd_casado}')
```

4. Entre homens e mulheres quem conseguiu mais empréstimos?

```python
df['sexo'].value_counts(normalize=True)
```

```python
df[df['aprovacao_emprestimo']=='Y']['sexo'].value_counts()
```

```python
#Solução 01
# Homens
filtro1 = df["sexo"] == "Male"
filtro2 = df["aprovacao_emprestimo"] == "Y"

num = len(df[filtro1 & filtro2])
percentual = num/len(df)
percentual = percentual*100
print(f"Homens: {round(percentual, 2)} %")

# Mulheres
filtro1 = df["sexo"] == "Female"
filtro2 = df["aprovacao_emprestimo"] == "Y"

num = len(df[filtro1 & filtro2])
percentual = num/len(df)
percentual = percentual*100
print(f"Mulheres: {round(percentual, 2)} %")
```

```python
#Solução 02
emprestimo = df['aprovacao_emprestimo'] == 'Y'
homens = df['sexo'] == 'Male'
mulheres = df['sexo'] == 'Female'
nulos = df['sexo'].isnull()

emprestimo_mulheres = df.loc[emprestimo & mulheres, 'cod_cliente'].count()
emprestimo_homens = df.loc[emprestimo & homens, 'cod_cliente'].count()
emprestimo_nulos = df.loc[emprestimo & nulos, 'cod_cliente'].count()
print(emprestimo_mulheres/614, emprestimo_homens/614, emprestimo_nulos/614)
```
