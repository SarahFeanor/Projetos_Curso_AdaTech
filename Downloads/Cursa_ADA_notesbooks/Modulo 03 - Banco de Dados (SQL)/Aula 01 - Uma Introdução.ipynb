# Banco de Dados: Uma Introdução

## O que vamos aprender?
 
Neste tópico vamos entender o que é um banco de dados, os tipos de banco de dados e como funciona o armazenamento no computador. Então, para isso vamos começar respondendo a primeira pergunta!
 
## O que é um banco de dados?
 
Em algum momento você já se perguntou onde são armazenadas as suas informações quando faz um cadastro em sites? Ou onde ficam guardadas as informações que você preenche na sua rede social? Como por exemplo posts, suas interações com publicações, seu histórico de pesquisa e todos os dados que temos acesso que ficam guardados como registros de atividade? 

Podemos pensar um pouco além e estender para a seguinte pergunta: Onde ficam armazenadas as informações que facilitam que os algoritmos das redes sociais identifiquem o que gostamos e tracem o nosso perfil, fazendo com que o aplicativo possa nos exibir publicações e vídeos que vão ao encontro às nossas preferências?
 
Tudo isso é possível através do **banco de dados**, que são justamente conjuntos de dados relacionados entre si, e quando organizados provêm uma informação como registros sobre pessoas, lugares ou comentários em posts.
 
Levando essas informações em consideração, podemos dizer então que um **banco de dados** é um software que ficará no computador, fazendo o gerenciamento dos dados que nossa aplicação precisa para ter utilidade. Um exemplo seria um sistema de e-commerce, neste caso precisamos guardar informações sobre produtos, clientes, pedidos, carrinho etc.
 
 
## Tipos de bancos de dados
 
Atualmente existem diversos tipos de bancos de dados, cada um refletindo objetivos diferentes. Podemos separar os bancos de dados em dois grandes grupos, sendo o primeiro deles o tipo de banco de dados que chamamos de **Relacional** (SQL), este é voltado para armazenar os dados em formato de tabelas que são conectadas entre si, e uma ótima vantagem é que esse tipo de banco de dados é muito parecido com o modelo do **Excel**, pois temos abas que contém as suas tabelas. Já o segundo grande grupo, chamamos de "**Não Relacional**" (NoSQL) em que o objetivo é armazenar dados em não tabular mas em formatos diversos como JSON, chave-valor, grafo, colunar etc.
 
Abaixo, podemos listar o nome de alguns bancos de dados relacionais como:
 
- PostgreSQL;
- Microsoft SQL Server (MSSQL);
- OracleDB;
- Mysql;
- SQLite;
- MariaDB.
 
E o nome de alguns bancos de dados não relacionais, como:
 
- MongoDb;
- Redis;
- Amazon DynamoDB;
- ElasticSearch;
- Cassandra;
- Neo4j.
 
## Como é feita a comunicação com um banco de dados?
 
Visto que temos diversos bancos de dados, precisamos entender como nos comunicamos com eles. Neste caso durante o módulo iremos utilizar o SQL Server que funciona como um servidor, o que significa que ele será instalado na nossa máquina, e a partir daí poderemos enviar comandos de uma linguagem chamada SQL (Structured Query Language) para o servidor, que resultará em uma ação do servidor no seu computador.
 
SQL é uma sigla para structured query language, ou, linguagem estruturada de consulta. Bancos de dados relacionais como o SQL Server, MySQL, MariaDB, por exemplo, utilizam essa linguagem para identificar o que precisa ser feito, então executamos comandos que o servidor está esperando para poder interpretar e realizar as operações no computador. 
 
Quando aprendemos a linguagem SQL utilizando qualquer servidor de banco de dados relacional, como SQL Server, podemos reaproveitar grande parte dos comandos para utilizá-los em outro servidor como MySQL ou MariaDB, as mudanças são muito pequenas, então acabamos aprendendo a linguagem uma vez só, para podermos aproveitar em outros servidores, sem precisar aprender do zero.
 
No início do tópico foi explicado que um banco de dados relacional funciona em formato de tabelas assim como no Excel, então vamos aprender a criar essas tabelas a partir dos comandos SQL, assim como aprenderemos a manipular (Criar, atualizar, remover ou consultar) os dados nessas tabelas a partir da linguagem SQL também.
 
No próximo tópico começaremos a configurar o ambiente local (sua máquina) para podermos avançar aprendendo a linguagem SQL e um pouco mais sobre servidores de banco de dados e sua utilização.
